```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def im2double(img):
    """
    Converts an image to double precision and normalizes it to the range [0, 1].
    This function mimics the behavior of MATLAB's im2double function.

    Args:
        img: The input image.

    Returns:
        The converted image as a double precision NumPy array.
    """
    min_val = img.min()
    max_val = img.max()
    if max_val - min_val == 0:  # Avoid division by zero if min and max are equal
        return img.astype(np.float64)
    else:
        return (img.astype(np.float64) - min_val) / (max_val - min_val)

I = cv2.imread('cat.jpg')
Id = im2double(I)
plt.imshow(Id)
plt.show()
```

```python
R = I[:, :, 0]
G = I[:, :, 1]
B = I[:, :, 2]

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

axs[0].imshow(R, cmap='Reds')
axs[0].set_title('R')

axs[1].imshow(G, cmap='Greens')
axs[1].set_title('G')

axs[2].imshow(B, cmap='Blues')
axs[2].set_title('B')

plt.tight_layout()
plt.show()
```

```python
from skimage.color import rgb2gray

Ipb = rgb2gray(I)
plt.imshow(Ipb, cmap='gray')
plt.show()
```
